                              ** Data Types and Structures Questions **
                              

### 1. What are data structures, and why are they important?

Data structures are ways of organizing and storing data so it can be accessed and modified efficiently.
They are important because they provide the foundation for designing efficient algorithms and writing optimized code.


### 2. Difference between mutable and immutable data types with examples

Mutable: Can be changed after creation (e.g., list, dict, set).
Immutable: Cannot be changed after creation (e.g., int, str, tuple).

Examples:
mutable_list = [1, 2, 3]
mutable_list.append(4) --> Modified

immutable_str = "hello"
immutable_str[0] = 'H' --> This will raise an error


### 3. Differences between lists and tuples

Lists are mutable, tuples are immutable.
Lists use more memory, tuples are faster.


### 4. How dictionaries store data

Dictionaries store data as key-value pairs using a hash table internally.


### 5. Why use a set instead of a list

Sets automatically remove duplicates and are faster for membership tests.


### 6. What is a string in Python, and how is it different from a list

A string is a sequence of characters. It is immutable, whereas lists are mutable.


### 7. How tuples ensure data integrity

Tuples are immutable, meaning once created, their data cannot be changed, ensuring integrity.


### 8. What is a hash table, and how does it relate to dictionaries

A hash table is a data structure that maps keys to values using a hash function.
Python's dictionaries use hash tables under the hood.


### 9. Can lists contain different data types?

Yes, lists in Python can store multiple data types.


### 10. Why are strings immutable in Python

To improve performance and security, strings are made immutable in Python.


### 11. Advantages of dictionaries over lists

Faster lookup, better for labeled data, no need to search through items by index.


### 12. When is tuple preferable over list?

When you want a fixed collection of data that should not be changed.
Example: Coordinates (x, y)


### 13. How sets handle duplicates

Sets automatically discard duplicate values.


### 14. How “in” works differently for lists vs. dictionaries

In lists: checks for value.
In dicts: checks for key.


### 15. Can you modify elements of a tuple?

No, tuples are immutable.


### 16. What is a nested dictionary?

A dictionary within another dictionary.
Example:
student = {'name': 'John', 'marks': {'math': 90, 'science': 80}}


### 17. Time complexity of accessing dict elements

O(1) average-case time complexity.


### 18. When are lists preferred over dictionaries?

When order matters or when dealing with a simple collection of items.


### 19. Why dictionaries are unordered (before Python 3.7)

They use hash functions to store keys; order is not guaranteed in older versions.


### 20. Difference between list and dictionary in data retrieval

List: access by index.
Dict: access by key.
